```python
import sys
import os

sys.path.append('../src/')

from read_data import DataObject
from constants import spend_categories
from dataframe_utilities import get_filter_index, subgroup_sum, upsample

OVS1619 = DataObject(source = "OVS", version = "1619")
TVAE = DataObject(source = "Synthetic", version = "?", model = "TVAE")
TGAN = DataObject(source = "Synthetic", version = "?", model = "TGAN")
CTGAN = DataObject(source = "Synthetic", version = "?", model = "CTGAN")
```

```python
assert all(col in list(OVS1619.data.columns) for col in list(TVAE.data.columns))
assert all(col in list(OVS1619.data.columns) for col in list(TGAN.data.columns))
assert all(col in list(OVS1619.data.columns) for col in list(CTGAN.data.columns))
```

```python
import math
import numpy as np

spend_cats = [cat+'_cat' for cat in spend_categories]

print(f"% of rows with at least one column of missing data")
print("---------------------------------------------------------------------")
for obj in (OVS1619, TVAE, TGAN, CTGAN):
    print(f"{obj}\n")
    invalid = (obj.data[spend_categories].isna() or obj.data[spend_categories] < 0).sum(axis = 1) >= 1
    print(f"{np.round(sum(invalid)/obj.size[0], 2)}")
    print("\n")
```

```python
print(f"% of complete rows")
print("-------------------")
for obj in (OVS1619, TVAE, TGAN, CTGAN):
    print(f"{obj}\n")
    positive_bool = (obj.data[spend_categories] >= 0).sum(axis = 1) == 5
    valid_bool = (obj.data[spend_cats] == 0).sum(axis = 1) == 5
    print(f"{np.round(sum(positive_bool + valid_bool)/obj.size[0], 2)}")
    print("\n")
```
