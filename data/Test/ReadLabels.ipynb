```python
import pandas as pd
import openpyxl

# Load the Excel file
wb = openpyxl.load_workbook("Data Dictionary for NUS_v5.xlsx")
sheet_name = "Variable Value"

sheet = wb[sheet_name]

# Initialize an empty dictionary to store the results
result = {}

# Initialize variables to track category and its values
category = None
category_values = {}

# Iterate over the rows in the sheet starting from row 3 to skip the first two rows
for row in sheet.iter_rows(min_row=3, max_row=sheet.max_row, min_col=1, max_col=3):
    cell_value = row[0].value
    
    # If the cell is not merged (assumes merged cells are meant as continuation of category)
    if row[0].value is not None:
        category = cell_value
        category_values = {}
        result[category] = category_values

    # Assuming columns B and C are not merged cells and contains the key-value pair
    if row[1].value is not None and row[2].value is not None:
        key = row[1].value
        value = row[2].value
        category_values[key] = value
```

```python
df = pd.read_csv("countries_and_nationality.csv", encoding='latin-1')
nation_map = df.set_index('Nationality')['Country'].to_dict()
result['NATIONALITY_TO_COUNTRY'] = nation_map

import json

json_filename = 'label_map.json'  # Define the JSON filename
with open(json_filename, 'w') as json_file:
    json.dump(result, json_file, indent=4)

print(f"Data saved to {json_filename}")

result
```
